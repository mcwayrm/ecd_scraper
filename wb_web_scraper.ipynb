```python
# Import Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import pandas as pd



# Creates a list of urls for each combo respectively
url = []
url.append('https://openknowledge.worldbank.org/discover?scope=%2F&query=%28%22early+childhood+development%22+OR+%22early+development%22+OR+%22early+child+care%22+OR+%22ecd%22%29+AND+%28%22literacy%22+OR+%22cognition%22+OR+%22education%22+OR+%22school*%22%29&submit=')
url.append('https://openknowledge.worldbank.org/discover?scope=%2F&query=%28%22early+childhood+development%22+OR+%22early+development%22+OR+%22early+child+care%22+OR+%22ecd%22%29+AND+%28%22heatlh*%22+OR+%22nutr*%22+OR+%22mental+health%22+OR+%22therapy%22+OR+%22special+needs%22+OR+%22food%22%29&submit=')
url.append('https://openknowledge.worldbank.org/discover?scope=%2F&query=%28%22early+childhood+development%22+OR+%22early+development%22+OR+%22early+child+care%22+OR+%22ecd%22%29+AND+%28%22women+empowerment%22+OR+%22working+mother*%22+OR+%22girl%22+OR+%22women%27s+rights%22+OR+%22gender*%22%29&submit=')
url.append('https://openknowledge.worldbank.org/discover?scope=%2F&query=%28%22early+childhood+development%22+OR+%22early+development%22+OR+%22early+child+care%22+OR+%22ecd%22%29+AND+%28%22income+supplement%22+OR+%22cash+transfer%22+OR+%22earn*%22+OR+%22income%22+OR+%22salar*%22%29&submit=')
url.append('https://openknowledge.worldbank.org/discover?scope=%2F&query=%28%22early+childhood+development%22+OR+%22early+development%22+OR+%22early+child+care%22+OR+%22ecd%22%29+AND+%28%22social-emotional%22+OR+%22physical%22+OR+%22employment%22+OR+%22family%22+OR+%22well-being%22+OR+%22well+being%22%29&submit=')


for k in range(0, len(url)):
    
    # Creates empty ElementLists for each variable
    report_title = []
    report_summary = []
    report_citation = []
    report_link = []
    report_publish_date = []
    report_author = []
    
    executable_path = {'executable_path': './chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless=False)
    
    browser.visit(url[k])
    response = requests.get(url[k])
    soup = BeautifulSoup(response.text, 'html.parser')

    # Creates a boolean variable that detects if the search page is the last page
    not_last_page = True
    
    # loops the search until the last page
    while(not_last_page):
        
        for i in range(0,len(browser.find_by_tag('h4')) - 10):
            try:
                browser.find_by_tag('h4')[i].click()

                # Creating the BeautifulSoup parser to extract the information from the report page
                html = browser.html
                reports_soup = BeautifulSoup(html, 'html.parser')
                report = reports_soup.select_one('div.main-content')

                # Extracting the information from the report page
                title = report.find('h2', class_='ds-div-head').get_text()
                summary = report.find('div', class_='okr-item-page-field-wrapper abstract').get_text().splitlines()
                citation = report.find('div', class_='citation').get_text().splitlines()
                link_path = report.find('div', class_='okr-item-page-field-wrapper uri').get_text()
                link = link_path[4:].splitlines()
                publish_date = report.find('div', class_='simple-item-view-other word-break').get_text()
                author = report.find('div', class_='authorprofile-item-view-link').get_text().splitlines()

                # Append the information into the empty lists
                report_title.append(title)
                report_summary.append(summary[1])
                report_citation.append(citation[1])
                report_link.append(link[1])
                report_publish_date.append(publish_date)
                report_author.append(author[1])

            except AttributeError as e:
                print(e)

            browser.back()
            
        # if the amount of articles in the search page is less than 10, this is indeed the last page
        if(len(browser.find_by_tag('h4')) != 20):
            not_last_page = False
            break
        
        # Create a click action to navigate to the next page
        browser.is_element_present_by_css('a[class="next-page-link"]', wait_time=2)
        next_page = browser.find_by_css('a[class="next-page-link"]')
        next_page[1].click()
    
    browser.quit()
    
    # Storing the variables into a dataframe
    data_df = pd.DataFrame(list(zip(report_title, report_summary, report_citation, report_link, report_publish_date, report_author)), 
                       columns=["title", "summary", "citation", "link", "publish_date", "author"])
    
    # Saving the dataframe into CSV
    data_df.to_csv("data_c%d.csv" %(k+1), index=False, encoding='utf-8')
```
